```python
## Create 100000x1 array from exponential distribution and plot histogram with uniform & normal
```

```python
exp_data = np.random.exponential(scale=1.0, size=(100000,))
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)
```

```python
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="red", label="Normal")
plt.hist(exp_data, density=True, bins=100, histtype="step", color="green", label="Exponential")
plt.axis([-2.5, 8, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
```

```python
plt.show()
```

```python
## Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5].
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
```

```python
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

```python
fig = plt.figure()
```

```python
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
```

```python
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

```python
##Calculate the pearson standard coefficient and spearman rank correlation for the [HP, Attack, Defense, Sp. Ark, Sp. Def, Speed] features in the seaborn tutorial.ipynb. Visualize the results using heatmaps. Make sure the values are displaced in the heatmap as well.
```

```python
import pandas as pd
```

```python
path1 = ('/content/drive/MyDrive/DL_lab1/Pokemon.csv')
```

```python
df = pd.read_csv(path1, encoding='latin1')
print(df.head())
```

```python
print(df.columns.tolist())
```

```python
df = df[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
```

```python
pearson_corr = df.corr(method='pearson')
spearman_corr = df.corr(method='spearman')
```

```python

fig, ax = plt.subplots(1, 2, figsize=(12,5))
sns.heatmap(pearson_corr, annot=True, cmap="coolwarm", ax=ax[0])
ax[0].set_title("Pearson Correlation")
sns.heatmap(spearman_corr, annot=True, cmap="coolwarm", ax=ax[1])
ax[1].set_title("Spearman Correlation")
plt.show()
```

```python
## PairGrid with modifications
```

```python
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)  # bivariate histogram
g.map_diag(sns.histplot, kde=True)  # hist + KDE on diagonal
g.map_lower(sns.kdeplot, fill=True)  # bivariate KDE on lower half
plt.show()
```
